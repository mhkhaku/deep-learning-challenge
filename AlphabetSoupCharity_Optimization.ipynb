```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

application_df.drop(columns=['EIN', 'NAME'], inplace=True)

application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_list = application_df.APPLICATION_TYPE.value_counts()

app_type_list
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = app_type_list.loc[app_type_list <= 500].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

class_list = application_df.CLASSIFICATION.value_counts()

class_list
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_list.loc[class_list > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

classifications_to_replace = class_list.loc[class_list <= 1000].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
converted_app_df = pd.get_dummies(application_df)

converted_app_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = converted_app_df.IS_SUCCESSFUL.values
X = converted_app_df.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Optimization changes: 
# 1. Added a third and fourth layer
# 2. Increased the nodes for the first two layers
# 3. Changed the activation functions 'swish'

number_input_features = len(X_train[0])
hidden_nodes_layer1 = 32
hidden_nodes_layer2 = 16
hidden_nodes_layer3 = 8
hidden_nodes_layer4 = 4

nn = tf.keras.models.Sequential()

# First hidden layer with L2 regularization
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation=tf.keras.layers.Activation('swish')))

# Second hidden layer with L2 regularization
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation=tf.keras.layers.Activation('swish')))

# Third hidden layer with L2 regularization
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation=tf.keras.layers.Activation('swish')))

# Fourth hidden layer with L2 regularization
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation=tf.keras.layers.Activation('swish')))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
# Optimization changes: Increase epochs to 50

fit_model = nn.fit(X_train_scaled, y_train, epochs=50, callbacks=[tf.keras.callbacks.ModelCheckpoint(filepath='h5_files_optimization\model_weights_optimization.h5', save_weights_only=True, save_freq=5)])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Optimized Loss: {model_loss}, Optimized Accuracy: {model_accuracy}")
```
